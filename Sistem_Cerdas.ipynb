```python
PNoDiscountNo = 5/9
PYesDiscountNo = 4/9

PNoDiscountYes = 12/41
PYesDiscountYes = 29/41

PNoDiscount = 17/50
PYesDiscount = 33/50
```

```python
pd.crosstab(df['Free Delivery'], df['Buy'])
```

```python
PNoDeliveryNo = 4/9
PYesDeliveryNo = 5/9

PNoDeliveryYes = 10/41
PYesDeliveryYes = 31/41

PNoDelivery = 14/50
PYesDelivery = 36/50
```

```python
# kondisi B dimana
# Day = Holiday
# Discount = Yes
# Free Delivery = Yes
# P(A) = No Buy = PNo
# P(B) = PYesDiscount, PYesDelivery, Pholiday
# P(B|A)1 = (PHoliday|No Buy) = PHolidayNo
# P(B|A)2 = (PYesDiscount|No Buy) = PYesDiscountNo
# P(B|A)3 = (PYesDelivery|No Buy) = PYesDeliveryNo
# P(B)1 = PHoliday
# P(B)2 = PYesDiscount
# P(B)1 = PYesDelivery

#Ditanyakan P(A|B)?
#Rumus
# P(A|B) = P(B|A)*P(A)/P(B)
```

```python
PNoBuy1 = ((PHolidayNo*PYesDiscountNo*PYesDeliveryNo)*PNo)/(PHoliday*PYesDiscount*PYesDelivery)
print(PNoBuy1)
```

```python
# Kondisi B dimana
# Day = Holiday
# Discount = Yes
# Free Delivery = Yes
# P(A) = Buy = PYes
# P(B) = PYesDiscount, PYesDelivery, PHoliday
# P(B|A)1 = (PHoliday|Buy) = PHolidayYes
# P(B|A)2 = (PYesDiscount|Buy) = PYesDiscountYes
# P(B|A)3 = (PYesDelivery|Buy) = PYesDeliveryYes
# P(B)1 = PHoliday
# P(B)2 = PYesDiscount
# P(B)3 = PYesDelivery
#Ditanyakan P(A|B) ?
# P(A|B) =  P(B|A)*P(A)/P(B)
```

```python
PBuy2 = ((PHolidayYes*PYesDiscountYes*PYesDeliveryYes)*PYes)/(PHoliday*PYesDiscount*PYesDelivery)
print(PBuy2)
```

```python
# Probability of No Buy = 0.12989982125784594
# Probability of Buy = 0.8440667832398943
```

```python
PSum = PNoBuy1 + PBuy2
print(PSum)
```

```python
PBuy = (PBuy2/PSum) *100
print(PBuy)
```

```python
PNoBuy = (PNoBuy1*PSum) *100
print(PNoBuy)
```
